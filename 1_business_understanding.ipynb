## Data Analysis Task

## Objectives

The 3 main objectives of the task are:
1. data exploration and understanding of the data structure and identification of the key input variables that determine the target.
2. training and testing models that predict the target value as well as possible from some or all of the available features.
3. generating a prediction for the dependent variable t1 in the unknown.csv dataset.

Other objectives include the following parts:
1. Data import: Loading the data from the csv files into a suitable format.
2. Exploration of the data: Using statistical or visualization tools necessary to understand the known data set (data set known.csv), considering the following:
    - Observing the relationship between the independent variables f1-6 and the dependent variable t1.
    - Considering whether adding additional attributes such as f2/f4 (f2 divided by f4) to the training data frame is useful.
    - Consideration of whether adding f6/f2 as an additional attribute is also useful.
    - Categorizing feature f5 into exactly 5 categories and adding it to the existing data frame as an additional feature f5_cat.
3. Clean up and preprocess: Based on the previous step, a decision is made whether further steps are
are necessary to prepare the data for modeling.
4. Modeling and model selection: Training of one or more models with the known data set:
    - Review how these models performed.
    - Review which model is most likely to be recommended to best represent the data set?
5. Prediction: Create a prediction for the unknown dataset (dataset unknown.csv) based on the final selected model. Saving the prediction as a .csv file and or .pkl file.
